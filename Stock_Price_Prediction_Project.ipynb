```python

#____import library___
import numpy as np
import pandas as pd
import plotly.graph_objects as go

#____read a data____
data = pd.read_csv("AAPL.csv")
data


```

```python
#_____visualize this stock price data__________

figure = go.Figure(data=[go.Candlestick(x=data["Date"], open=data["Open"], high=data["High"], low=data["Low"], close=data["Close"])])
figure.update_layout(title = "Apple Stock Price Analysis", xaxis_rangeslider_visible = False)
figure.show()
```

```python
#______install autots for working with time series____
pip install autots
```

```python
#_________stock Prediction model______
from autots import AutoTS
model = AutoTS(forecast_length=7, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
predictions = model.predict()
forecast = predictions.forecast
```

```python
#_______result of stock market prediction for next week________
print(forecast)
```
